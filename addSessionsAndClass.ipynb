# Ajouter les sessions aux .csv
0 : Matin
1 : Soir

```python
import os
import pandas as pd

folders = ["FATIGUE_EVENING", "FATIGUE_MORNING", "NORMAL_EVENING", "NORMAL_MORNING"]

for folder in folders:
    files = os.listdir(folder)
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(folder, file)
            df = pd.read_csv(file_path)
            session_value = 1 if "EVENING" in folder else 0
            df['session'] = session_value
            df.to_csv(file_path, index=False)
            print(f"Session column added and set for file: {file}")

print("All files processed successfully.")
```

# Ajout des classes

1 : Fatigue
